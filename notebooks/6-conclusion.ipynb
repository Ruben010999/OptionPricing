## 2. Black–Scholes Model: Structure and Assumptions

The Black–Scholes model assumes that the underlying follows a geometric Brownian motion:

$$
dS_t = \mu S_t dt + \sigma S_t dW_t
$$

Under risk-neutral valuation:

$$
dS_t = r S_t dt + \sigma S_t dW_t
$$

Key assumptions:

- Constant volatility $\sigma$
- Constant risk-free rate $r$
- No transaction costs
- Continuous trading
- No jumps
- Lognormal price distribution

The European call price is:

$$
C(S_0, K, T) = S_0 N(d_1) - K e^{-rT} N(d_2)
$$

where

$$
d_1 = \frac{\ln(S_0/K) + (r + \tfrac{1}{2}\sigma^2)T}{\sigma \sqrt{T}}, \quad
d_2 = d_1 - \sigma \sqrt{T}
$$

This model provided the theoretical backbone for all subsequent analysis.

## 3. Greeks and Risk Decomposition

We computed and analyzed:

- Delta: $\Delta = \frac{\partial V}{\partial S}$
- Gamma: $\Gamma = \frac{\partial^2 V}{\partial S^2}$
- Vega: $\nu = \frac{\partial V}{\partial \sigma}$
- Theta: $\Theta = \frac{\partial V}{\partial t}$

The Greeks allow local risk decomposition:

$$
dV \approx \Delta dS + \frac{1}{2}\Gamma dS^2 + \Theta dt
$$

This decomposition is fundamental for:

- Hedging  
- Market making  
- Volatility trading  
- Risk management

## 5. Dynamic Hedging and Gamma Scalping

We implemented a discrete delta-hedging engine and simulated geometric Brownian motion paths.

For a delta-hedged portfolio:

$$
\Pi_t = V_t - \Delta_t S_t
$$

Under continuous hedging, Black–Scholes implies perfect replication.  
Under discrete hedging, P&L emerges from realized volatility differences.

We demonstrated:

- If realized volatility $\sigma_{real} > \sigma_{imp}$,  
  a long gamma (long straddle) strategy tends to generate positive P&L.
- If realized volatility $\sigma_{real} < \sigma_{imp}$,  
  the same strategy tends to generate negative P&L.

This confirms the theoretical relationship:

$$
\mathbb{E}[P\&L] \propto \sigma_{imp}^2 - \sigma_{real}^2
$$

---

## 6. Volatility Arbitrage (Monte Carlo Evidence)

To move beyond single-path illustrations, we conducted many-path Monte Carlo simulations.

We examined delta-hedged short straddles under two regimes:

1. $\sigma_{imp} > \sigma_{real}$  
   → Positive expected P&L  

2. $\sigma_{imp} < \sigma_{real}$  
   → Negative expected P&L  

The distributional analysis confirmed that volatility trading is fundamentally a relative-value trade between implied and realized volatility.

This establishes the link between:

- Black–Scholes pricing,
- Greeks,
- Dynamic hedging,
- Practical volatility arbitrage.

## 7. Limitations of the Framework

Despite its elegance, the Black–Scholes model has significant limitations:

1. Volatility is not constant.  
2. Markets exhibit jumps.  
3. Volatility surfaces display skew and smile.  
4. Hedging is discrete and costly.  
5. Transaction costs and liquidity constraints matter.  

In practice:

- Realized volatility is stochastic.  
- Implied volatility varies across strikes and maturities.  
- Tail risk is underrepresented by lognormal dynamics.  

Therefore, the Black–Scholes model should be viewed as a foundational approximation rather than a complete market description.

## 9. Final Remarks

This project was designed not merely as a coding exercise, but as a full conceptual journey through quantitative option theory — from mathematical foundations to trading intuition and simulated P&L.

We began with the Black–Scholes differential equation and its closed-form solution, establishing the theoretical pricing structure under risk-neutral valuation. From there, we translated analytical formulas into numerical implementations, ensuring that every component — $d_1$, $d_2$, option value, and Greeks — was fully understood both mathematically and computationally.

We then moved from static valuation to risk decomposition. The Greeks transformed the pricing formula into a dynamic sensitivity framework. Instead of viewing an option as a static object, we interpreted it as a bundle of risk exposures:

- $\Delta$ — directional exposure  
- $\Gamma$ — convexity exposure  
- $\Theta$ — time decay  
- $\nu$ — volatility exposure  

This naturally led to the next conceptual step: hedging.

By implementing a discrete delta-hedging engine and simulating geometric Brownian motion paths, we bridged the gap between theoretical replication and practical execution. The project clearly demonstrated that:

- Continuous hedging is a theoretical idealization.
- Discrete hedging introduces residual P&L.
- That residual P&L is fundamentally driven by the relationship between implied and realized volatility.

The gamma scalping experiments provided the key economic insight:

$$
\text{Volatility trading is a bet on } \sigma_{real} \text{ relative to } \sigma_{imp}.
$$

Through single-path and many-path Monte Carlo simulations, we observed:

- Long gamma profits when realized volatility exceeds implied volatility.
- Short gamma profits when implied volatility exceeds realized volatility.
- The distribution of P&L reflects volatility mispricing rather than directional forecasting.

This is a central principle of modern derivatives markets:  
options are instruments for structuring and transferring volatility risk.

Importantly, the project also clarified the boundaries of the Black–Scholes world. Real markets exhibit:

- Volatility smiles and skews,
- Stochastic volatility,
- Jumps,
- Liquidity frictions,
- Transaction costs.

Thus, while the model is internally consistent and mathematically elegant, it is a first-order approximation of reality.

From a broader perspective, this project unified:

- Stochastic calculus,
- Risk-neutral pricing,
- Numerical simulation,
- Strategy construction,
- Hedging mechanics,
- Statistical analysis of P&L distributions.

It demonstrated how mathematical theory translates into trading logic, and how pricing models connect directly to risk management and volatility arbitrage.

In summary, the project achieved three major goals:

1. Built a complete Black–Scholes pricing and Greeks engine.

The final outcome is a coherent quantitative framework that links theory, computation, and market intuition — forming a strong foundation for further exploration of advanced derivative models and real-world quantitative trading systems.
